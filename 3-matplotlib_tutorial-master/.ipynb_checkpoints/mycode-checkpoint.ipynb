```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas.describe()
```

```python
columns = gas.columns
```

```python
plt.figure(figsize=(8, 5))

plt.title('Gas Prices Over Time(in USD)', fontdict={'fontweight':'bold', 'fontsize':22})

for column in columns[1:]:
    plt.plot(gas[columns[0]], gas[column], label=column, marker='.')
    
plt.xticks(gas.Year, rotation=45)

plt.xlabel('Year')
plt.ylabel('US Dollars')
    
plt.legend()

plt.savefig("GAS.png", dpi=300)

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
```

```python
fifa.head()
```

### Histogram

```python
bins = list(np.arange(40, 110, 10))

plt.hist(fifa.Overall, bins=bins, color='#c752a2')

plt.xticks(bins)

plt.xlabel('skill level')
plt.ylabel('Number of players')
plt.title('Distribution of Player in FIFA 2018')

plt.show()
```

```python
plt.pie()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
```

```python
labels = ['left', 'right']

plt.pie([left,right], labels=labels,autopct='%.2f %%')

plt.show()
```

```python
fifa.head()
```

```python
fifa.Weight = [int(i.strip('lbs')) if type(i)==str else i for i in fifa.Weight]
```

```python
plt.style.use('ggplot')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight > 200].count()[0]
weight = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125~150', '150~175', '175~200', 'Over 200']
explode = [0.5, 0.2, 0, 0, 0.5]
plt.title("fifa player weight")
plt.pie(x=weight, labels=labels, autopct='%.2f %%', pctdistance=0.8, explode=explode)
plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
rev = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
```

```python
plt.style.use('default')
plt.figure(figsize=(3, 5))
labels = ['FC Barcelona', 'Real Madrid', 'rev']
boxes = plt.boxplot([barcelona, madrid, rev], labels=labels)

plt.title('FIFA Team')
plt.ylabel('Overall rating')
plt.show()
```
