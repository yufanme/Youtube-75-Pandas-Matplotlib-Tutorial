### Baisic Graph

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
# plt.plot(x, y, label='2x', color='r', linewidth=1, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# resize your graph
plt.figure(figsize=(10, 6), dpi=300)


# quick 
plt.plot(x, y, 'b^--', label='2x',)

# line number 2
x2 = np.arange(0, 4.5, 0.5)
print(x2)

plt.plot(x2[:5], x2[:5]**2, color='r', label='x^2')
plt.plot(x2[4:], x2[4:]**2, 'r--', label='x^2')

plt.title(label="Our First Graph", fontdict={'fontsize':20, 'fontname': 'Comic Sans MS'})
plt.xlabel(xlabel='X Axis!', labelpad=20, loc='right')
plt.ylabel(ylabel='Y Axis!', labelpad=20, loc='top')
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])
plt.legend()
plt.savefig('mygraph.png', dpi=300)
plt.show()
```

### bar chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]
bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.figure(figsize=(6, 4))

plt.show()
```
